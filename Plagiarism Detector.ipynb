## Firstly, We have defined a function which will calculate the levenshtein distance of the Test and Source sentences.

```python
def leven_distance(s1, s2):
  m=len(s1)+1
  n=len(s2)+1
  mat = {}
  for i in range(m): 
    mat[i,0]=i
  for j in range(n): 
    mat[0,j]=j
  #print(tbl)
  for i in range(1, m):
    for j in range(1, n):
      if s1[i-1] == s2[j-1]:
        cost = 0 
      else:
        cost = 2
      mat[i,j] = min(mat[i, j-1]+1, mat[i-1, j]+1, mat[i-1,j-1]+cost)
  return mat[i,j]
```

```python
def rem_stop_words(line,stopwords):
  #convert line to lowercase, split wordwise
  line_tokens = line.lower().split()
  #make a list for the word not in the stopword list
  tokens_without_sw = [word for word in line_tokens if not word in stopwords]
  new_line = " ".join(tokens_without_sw)
  return new_line
```

## Here we have defined a function which will split the sentence line by line.
## We will be calculating the plagarism for each line. Accordingly the plagarism percentage will be calculated.
## This step can also be referred to as the tokenisation of sentences.

```python
def para_split(line):
  line_split = line.split(". ")
  #print(line_split)
  return line_split
```

## We will now be calling all the functions defined above for the pre-processing of the sentences. After which we will be calculating the levenshtein distance.
## Here, whenever the distance between the Test and Source is less than that of the length of the Test sentence by 2, then we say that the sentence is plagiarised.
## The count of the number of sentences plagiarised is updated everytime the condition for the same is satisfied.
## According to the count, the plagiarism percentage is calculated.

```python
#without nltk
def main(Test,Source):
  print("The test sentence against which plagrarism is to be detected:\n",Test)
  print("The source sentence whose plagrarism is to be detected:\n",Source)
  num_of_plag_line = 0
  TestArr = []
  SourceArr = []
  new_Test = rem_stop_words(Test,stopwords)
  new_Source = rem_stop_words(Source,stopwords)
  print("The test sentence after removal of stopwords is:\n",new_Test)
  print("The source sentence after removal of stopwords is:\n",new_Source)

  for s in para_split(new_Test) :
    TestArr.append(s)
  for s in para_split(new_Source) :
    SourceArr.append(s)
  count = 1
  for s1,s2 in zip(TestArr,SourceArr) :
    #Condition for plagarism
    if leven_distance(s1,s2) < len(s1)/2:
      print("Sentence {} is plagaraised".format(count))
      num_of_plag_line = num_of_plag_line+1
    count = count+1
  if num_of_plag_line == 0:
      print("The document is 100% unique")
  else:
    print("The document has {} % of plagrarism detected".format((num_of_plag_line/len(SourceArr))*100))
```
